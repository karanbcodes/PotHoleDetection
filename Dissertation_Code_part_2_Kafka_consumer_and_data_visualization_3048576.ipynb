# Importing packages of Kafka-python for consumer

```python
from kafka import KafkaConsumer
from time import sleep
import json 
from datetime import datetime
from json import loads
import pandas as pd
```

# Kafka Consumer for Printing the messages to console in real-time

```python
# # commenting this out because we need to push the data to dataframe
# we can use this code to check how consumer id consuming data in real-time
# this terminates automatically when there is no more data to be read
consumer = KafkaConsumer('pothole-location', 
                         bootstrap_servers=['localhost:9092'], 
                         api_version=(0, 10) 
#                        this will check the consumer for 4000ms and will close if no new msg is there
                         ,consumer_timeout_ms=6000
                        )
for message in consumer:
    print(json.loads(message.value))
KafkaConsumer.close()
```

# Kafka consumer for Writing the messages to dataframe in real-time

```python
# this terminates automatically when all the data has been ingested
consumer = KafkaConsumer('pothole-location', 
                         bootstrap_servers=['localhost:9092'], 
                         api_version=(0, 10) 
#                        this will check the consumer for 4000ms and will close if no new msg is there
                         ,consumer_timeout_ms=6000
                        )
df_coordinates = pd.DataFrame()
for message in consumer:
#     deserializing the recieved data here
    dict_new = json.loads(message.value)
    df_coordinates = df_coordinates.append(dict_new, ignore_index=True)
KafkaConsumer.close()
```

```python
# Creating an OpenStreetMap
map = folium.Map(location = geocoder.ip('me').latlng, tiles='OpenStreetMap' , zoom_start = 20)
map
```

# Plotting the predicted and streamed locations of potholes

```python
for index, location_info in df_coordinates.iterrows():
    folium.Marker([location_info["Lat"], location_info["Long"]]).add_to(map)
```
